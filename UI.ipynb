```python
import tkinter as tk
from tkinter import Label, Button
from PIL import Image, ImageTk
import cv2
import random
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

import import_ipynb
import lib_Processdata

# เส้นทางภาพ
image_path = r"ui1.jpg"

# โหลดภาพต้นฉบับ
original_image = cv2.imread(image_path)
if original_image is None:
    raise FileNotFoundError(f"ไม่พบไฟล์: {image_path}")

# แปลงภาพเป็นสีเทา
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

# สร้างหน้าต่าง GUI
window = tk.Tk()
window.title("Trigger Point Detector")
window.geometry("1920x1080")
####Full HD

# ตัวแปรสถานะ
highlight = tk.BooleanVar(value=False)
smoothness = lib_Processdata.cal_Smoothness_signal_ch0
variance = lib_Processdata.cal_variance_signal_ch0
variance = (variance/0.00500) *100
variance = round(variance, 4)  # ปัดทศนิยมให้มี 2 ตำแหน่ง
print(variance)

# variance = 1
# ฟังก์ชันสำหรับแสดงภาพ
def display_image():
    if variance > 0.00500 :
        image_to_display = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
    else:
        image_to_display = cv2.cvtColor(gray_image, cv2.COLOR_GRAY2RGB)
    # แปลงภาพสำหรับ tkinter
    image = Image.fromarray(image_to_display)
    image = image.resize((250, 250))  # ปรับขนาดภาพ
    img_tk = ImageTk.PhotoImage(image)
    image_label.configure(image=img_tk)
    image_label.image = img_tk

# ฟังก์ชันสำหรับเปลี่ยนข้อความ
def update_text():
    if variance < 0.00500 :
        status_label.config(text="TRIGGERPOINT WAS FOUND", fg="red")
        graph_label.config(bg="red")
    else:
        status_label.config(text="TRIGGERPOINT WAS NOT FOUND", fg="gray")
        graph_label.config(bg="gray")

# ส่วนแสดงภาพ
image_label = Label(window, bg="white")
image_label.place(x=100, y=10)  # จัดตำแหน่งภาพ

# ส่วนแสดงข้อความสถานะ
if variance < 0.00500 :
    status_label = Label(window, text="TRIGGERPOINT WAS NOT FOUND", font=("Calibri", 40), fg="gray")
    status_label.place(x=400, y=70)  # วางข้อความใต้ภาพ
else:
    status_label = Label(window, text="TRIGGERPOINT WAS FOUND", font=("Calibri", 40), fg="red")
    status_label.place(x=400, y=70)

# ส่วนแสดงค่า smoothness
answer_var = tk.StringVar()
answer_var.set(str(smoothness))  # ตั้งค่าเริ่มต้นให้กับ answer_var
# สร้าง label ที่จะแสดงค่า answer
status_label = Label(window, textvariable=answer_var, font=("Calibri", 30), fg="black")
status_label.place(x=660, y=130)  # วางข้อความใต้ภาพ

# ส่วนแสดงค่า variance
answer_var = tk.StringVar()
answer_var.set(f"{variance:.1f}%")
# สร้าง label ที่จะแสดงค่า answer
status_label = Label(window, textvariable= answer_var, font=("Calibri", 30), fg="black")
status_label.place(x=1220, y=130)  # วางข้อความใต้ภาพ

# ส่วนแสดงค่า smoothness และ variance
smoothness_label = Label(window, text="SMOOTHNESS: ", font=("Calibri", 30))
smoothness_label.place(x=400, y=130)
variance_label = Label(window, text="VARIANCE: ", font=("Calibri", 30))
variance_label.place(x=1030, y=130)

graph_image = Image.open('output_graph.png')  # โหลดภาพจากไฟล์ที่บันทึก
graph_image = graph_image.resize((1050, 470))  # ปรับขนาดภาพกราฟ
graph_img_tk = ImageTk.PhotoImage(graph_image)
graph_label = Label(window, image=graph_img_tk)
graph_label.place(x=230, y=290)  # วางตำแหน่งกราฟ

# เริ่มแสดงภาพครั้งแรก
display_image()

# เริ่มต้นหน้าต่าง GUI
window.mainloop()
```
